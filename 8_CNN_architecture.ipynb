```python
import tensorflow
import keras
from keras import Sequential
from keras.layers import Conv2D , MaxPooling2D , Dense , Flatten, GlobalAveragePooling2D
```

```python

model   = Sequential()
model.add(Conv2D(32,3,padding="same",activation="relu",input_shape=(256,256,3)))
model.add(MaxPooling2D(pool_size=(2,2),strides=2))
model.add(Conv2D(64,3,padding="same",activation="relu"))
model.add(MaxPooling2D(pool_size=(2,2),strides=2))
model.add(Conv2D(128,3,padding="same",activation="relu"))
model.add(MaxPooling2D(pool_size=(2,2),strides=2))
model.add(Conv2D(256,3,padding="same",activation="relu"))
model.add(MaxPooling2D(pool_size=(2,2),strides=2))

model.add(Flatten())
model.add(Dense(128,activation="relu"))
model.add(Dense(2,activation="softmax"))


```

```python
# ==========================================
# CNN ARCHITECTURE AND WORKING FLOW
# ==========================================

# Q1) CNN kya hota hai?
# Answer:
# CNN (Convolutional Neural Network) ek Deep Learning model hai
# jo images ko process karne ke liye use hota hai.
# CNN image ke features jaise edges, shapes, patterns detect karta hai.


# ==========================================
# CNN WORKING FLOW (STEP BY STEP)
# ==========================================

# Step 1) Input Image
# Example:
# Image shape = (28,28,3)
# Matlab:
# 28 = Height
# 28 = Width
# 3 = RGB channels

# Computer image ko matrix ke form me dekhta hai.

# Example:
# images[0].shape = (28,28,3)


# ==========================================
# Step 2) Convolution Operation
# ==========================================

# CNN me sabse pehle convolution operation apply hota hai.

# Example:
# Filter size = 3x3

# Filter image ke upar move karta hai.

# Har movement par dot product hota hai.

# Result:
# Feature map generate hota hai.

# Example:
# Input = (28,28,3)
# Filter = (3,3,3)

# Output ≈ (26,26)


# ==========================================
# Step 3) Activation Function
# ==========================================

# Convolution ke baad activation function apply hota hai.

# Mostly ReLU use hota hai.

# ReLU formula:

# f(x) = max(0,x)

# Negative values remove ho jati hain.

# Model nonlinear ban jata hai.


# ==========================================
# Step 4) Padding
# ==========================================

# Padding image ke border par zeros add karta hai.

# Purpose:
# Image ka size kam hone se bachana.

# Example:

# Without Padding:
# 28 -> 26 -> 24 -> 22 (size fast reduce hoga)

# With Padding:
# 28 -> 28 -> 28


# ==========================================
# Step 5) Max Pooling (Cleanup Process)
# ==========================================

# Max pooling image ka size reduce karta hai.

# Example:

# Input = 28x28
# Pooling = 2x2

# Output = 14x14

# Max pooling important features rakhta hai
# aur noise remove karta hai.

# Isko cleanup process bhi bol sakte hain.


# ==========================================
# Step 6) Convolution Repeat
# ==========================================

# CNN me convolution ek hi baar nahi hota.

# Multiple layers hoti hain:

# Conv → ReLU → Pooling
# Conv → ReLU → Pooling
# Conv → ReLU → Pooling

# Har layer deeper features detect karti hai.

# First Layer:
# Edges detect karta hai

# Second Layer:
# Shapes detect karta hai

# Third Layer:
# Objects detect karta hai


# ==========================================
# Step 7) Feature Maps
# ==========================================

# Har convolution layer multiple filters use karti hai.

# Example:

# 32 filters → 32 feature maps
# 64 filters → 64 feature maps

# Matlab:

# Conv2D(32)

# Output:

# (Height,Width,32)


# ==========================================
# Step 8) Flatten Layer
# ==========================================

# Convolution ke baad output 3D hota hai.

# Example:

# (7,7,64)

# Flatten layer ise 1D vector me convert karta hai.

# Example:

# (7×7×64)

# Output:

# 3136


# ==========================================
# Step 9) Dense Layer
# ==========================================

# Flatten ke baad Dense layer lagate hain.

# Dense layer fully connected layer hoti hai.

# Ye final decision leta hai.


# Example:

# Dense(128)

# Matlab:

# 128 neurons


# ==========================================
# Step 10) Output Layer
# ==========================================

# Final layer prediction karti hai.

# Example:

# Cat vs Dog

# Dense(1) → Sigmoid

# Output:

# 0 = Cat
# 1 = Dog


# ==========================================
# Step 11) CNN Complete Flow
# ==========================================

# Image Input
# ↓
# Convolution
# ↓
# ReLU
# ↓
# Padding
# ↓
# Max Pooling
# ↓
# Convolution
# ↓
# ReLU
# ↓
# Max Pooling
# ↓
# Flatten
# ↓
# Dense
# ↓
# Output


# ==========================================
# Step 12) CNN Kab Tak Apply Karte Hain?
# ==========================================

# Convolution layers tab tak lagate hain
# jab tak image ke important features extract ho jaye.

# Usually:

# 2–5 Conv layers normal hote hain.


# ==========================================
# Step 13) CNN Model Example
# ==========================================

# Input (28,28,3)
# ↓
# Conv2D(32)
# ↓
# MaxPooling
# ↓
# Conv2D(64)
# ↓
# MaxPooling
# ↓
# Flatten
# ↓
# Dense(128)
# ↓
# Dense(1)


# ==========================================
# Step 14) CNN Main Idea
# ==========================================

# Image
# ↓
# Features Extract
# ↓
# Important Features Select
# ↓
# Classification


# CNN image ko directly classify nahi karta.
# Pehle features nikalta hai
# phir decision leta hai.
```
